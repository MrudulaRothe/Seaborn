# Categorical Plots
Seaborn also helps us in doing the analysis on Categorical data points
- count plot
- bar plot
- box plot
- cat plot
- Violin plot
- Strip plot
- Swarm plot

### Countplot
In this plot we have to give either x value or y value.
Other one is always count

```python
sns.countplot('day',data=df) # it is by default x='day'
```

```python
sns.countplot(y='sex',data=df)
```

## Bar plot
Here we have to give both x and y values

```python
sns.barplot(y='total_bill',x='smoker',data=df,estimator=np.median) #np.mean,np.median,np.cov,np.var
```

## Box plot
- It gives the 5 number summary

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df, palette='Blues')
```

```python
# If we do not specify x and y, Boxplot will by default take all numerical features
sns.boxplot(data=df,orient='v') # v=vertical orientation h=horizontal orientation
```

```python
sns.boxplot(x='total_bill',y='day',data=df,hue='smoker')
plt.legend(loc=0)
```

## Cat plot

```python
tips=sns.load_dataset('tips')
sns.catplot(x='day', y='total_bill', kind='boxen', data=tips)
```

## ViolinPlot
Violin Plot helps us to see both the distribution in terms of Kernal density estimation and the box plot

```python
sns.violinplot(x='day',y='total_bill', data = df, hue = 'sex', split = True)
```

## Strip plots

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=df, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=df, jitter=True, hue='sex', dodge= True)
```

## Swarm Plot
Its combination of violinplot and stripplot

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='day',y='total_bill', data = df, hue = 'sex', dodge = True)
sns.swarmplot(x='day',y='total_bill', data = df, hue = 'sex', dodge = True, color='white')

# To locate legend out of the grid 
plt.legend(bbox_to_anchor=(1, 1))
```

```python
sns.swarmplot(x='day',y='total_bill', data = df, hue = 'sex')
```

## Palettes

```python
plt.figure(figsize=(8,8))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=df, jitter=True, hue='sex', palette='magma')
# sns.stripplot(x='day',y='total_bill',data=df, jitter=True, hue='sex', palette='afmhot')

# Location of legend
# upper right:1, upper left: 2, lower left: 3, lower right: 4
plt.legend(loc=1)

# To locate legend out of the grid 
plt.legend(bbox_to_anchor=(1, 1))
```

## HeatMaps

```python
crash_df=sns.load_dataset("car_crashes")
crash_df.head()
```

```python
plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale=1.4)

crash_mx=crash_df.corr()

sns.heatmap(crash_mx, annot=True, cmap='Blues') # annot= annotations i.e numbers in the box
```

### Using pivot table

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(flights,cmap='Blues',linecolor='white',linewidth=1)
```

## Cluster Maps

```python
iris=sns.load_dataset('iris')
species=iris.pop('species')
sns.clustermap(iris)
```

```python
sns.clustermap(flights,cmap='Blues',standard_scale=1)
```

## PairGrid

```python
iris=sns.load_dataset('iris')
grid=sns.PairGrid(iris,hue='species')
grid.map_diag(plt.hist)
grid.map_offdiag(plt.scatter)
```

```python
grid=sns.PairGrid(iris,hue='species')
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
grid.map_diag(plt.hist)
```

```python
# Customizing the grid
grid=sns.PairGrid(iris,hue='species',
                 x_vars=['sepal_length','sepal_width'],
                 y_vars=['petal_length','petal_width'])
grid.map_diag(plt.hist)
grid.map_offdiag(plt.scatter)
grid.add_legend()
```

## Facet Grid

```python
tips_df=sns.load_dataset('tips')

tips_fg=sns.FacetGrid(tips_df,col='time',row='smoker')
tips_fg.map(plt.hist,'total_bill',bins=10)
```

```python
tips_fg=sns.FacetGrid(tips_df,col='time',row='smoker')
tips_fg.map(plt.scatter,'total_bill', 'tip')
```

```python
tips_fg=sns.FacetGrid(tips_df,col='time',hue='smoker', height=4 , aspect=1.3 ,
                     col_order=['Dinner','Lunch'],palette='Set1')
tips_fg.map(plt.scatter,'total_bill', 'tip',edgecolor='black')
```

```python
kws = dict(s=50,linewidth=0.5, edgecolor = 'w')
tips_fg=sns.FacetGrid(tips_df,col='sex',hue='smoker', height=4 , aspect=1.3 ,
                     hue_order=['Yes','No'],
                     hue_kws= dict(marker=['^','v']))
tips_fg.map(plt.scatter,'total_bill', 'tip', **kws)
```

## Regression Plots

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df, markers=['o','^'],palette='magma',
          scatter_kws={'s':80, 'linewidth':0.5 })
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
sns.lmplot(x='total_bill', y='tip', col='sex', row='time', data=tips_df)
```
